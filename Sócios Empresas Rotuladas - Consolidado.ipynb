```python
import pandas as pd
import numpy as np
```

```python
# IMPORTAÇÃO DA BASE DE SÓCIOS
socios_empresas_rotuladas_df = pd.read_csv(r'C:\DATASETS\Empresas Rotuladas\socios_empresas_rotuladas.csv', dtype=str, encoding='UTF-8')
socios_empresas_rotuladas_df
```

### Junções

```python
# Junção com a base de sócios que solicitaram o Auxílio Emergencial:
df = pd.read_csv(r'C:\DATASETS\Empresas Rotuladas\socios_empresas_rotuladas-auxilio_emergencial.csv', dtype=str, encoding='UTF-8')
socios_empresas_rotuladas_df = socios_empresas_rotuladas_df.merge(df, on=['cnpj', 'cnpj_cpf_socio', 'nome_socio', 'cnpj_cpf_socio_mascarado'], how='left')
# auxilio_emergencial_bloqueado, socio_solicitou_auxilio_emergencial
```

```python
# Junção com a base de Sócios que solicitaram Bolsa Família:
df = pd.read_csv(r'C:\DATASETS\Empresas Rotuladas\socios_empresas_rotuladas-bolsa_familia.csv', dtype=str, encoding='UTF-8')
socios_empresas_rotuladas_df = socios_empresas_rotuladas_df.merge(df, on=['cnpj', 'cnpj_cpf_socio', 'nome_socio', 'cnpj_cpf_socio_mascarado'], how='left')
# socio_solicitou_bolsa_familia
```

```python
# Junção com a base de Sócios Candidatos:
df = pd.read_csv(r'C:\DATASETS\Empresas Rotuladas\socios_empresas_rotuladas-candidato.csv', dtype=str, encoding='UTF-8')
socios_empresas_rotuladas_df = socios_empresas_rotuladas_df.merge(df, on=['cnpj', 'cnpj_cpf_socio', 'nome_socio', 'cnpj_cpf_socio_mascarado'], how='left')
# socio_candidato
```

```python
# Junção com a base de Sócios Devedores da União:
df = pd.read_csv(r'C:\DATASETS\Empresas Rotuladas\socios_empresas_rotuladas-devedores_da_uniao.csv', dtype=str, encoding='UTF-8')
socios_empresas_rotuladas_df = socios_empresas_rotuladas_df.merge(df, on=['cnpj', 'cnpj_cpf_socio', 'nome_socio', 'cnpj_cpf_socio_mascarado'], how='left')
# socio_devedor_uniao
```

```python
# Junção com a base de Sócios que Participam de Empresa Inexistente:
df = pd.read_csv(r'C:\DATASETS\Empresas Rotuladas\socios_empresas_rotuladas-participa_empresa_inexistente.csv', dtype=str, encoding='UTF-8')
socios_empresas_rotuladas_df = socios_empresas_rotuladas_df.merge(df, on=['cnpj', 'cnpj_cpf_socio', 'nome_socio', 'cnpj_cpf_socio_mascarado'], how='left')
# empresas_inexistentes_socio_participa
```

```python
# Junção com a base de Sócios que Participam de outras Empresas:
df = pd.read_csv(r'C:\DATASETS\Empresas Rotuladas\socios_empresas_rotuladas-participa_outros_cnpjs.csv', dtype=str, encoding='UTF-8')
socios_empresas_rotuladas_df = socios_empresas_rotuladas_df.merge(df, on=['cnpj', 'cnpj_cpf_socio', 'nome_socio', 'cnpj_cpf_socio_mascarado'], how='left')
# qte_cnpj_socio_participa
```

```python
# Junção com a base de Sócios que são Pessoas Politicamente Expostas:
df = pd.read_csv(r'C:\DATASETS\Empresas Rotuladas\socios_empresas_rotuladas-pessoa_politicamente_exposta.csv', dtype=str, encoding='UTF-8')
socios_empresas_rotuladas_df = socios_empresas_rotuladas_df.merge(df, on=['cnpj', 'cnpj_cpf_socio', 'nome_socio', 'cnpj_cpf_socio_mascarado'], how='left')
# socio_pessoa_politicamente_exposta
```

```python
# Junção com a base de Sócios que receberam recursos do Governo Federal:
df = pd.read_csv(r'C:\DATASETS\Empresas Rotuladas\socios_empresas_rotuladas-recebimento_recurso_governo_federal.csv', dtype=str, encoding='UTF-8')
socios_empresas_rotuladas_df = socios_empresas_rotuladas_df.merge(df, on=['cnpj', 'cnpj_cpf_socio', 'nome_socio', 'cnpj_cpf_socio_mascarado'], how='left')
# socio_recurso_recebido_governo_federal
```

```python
# Junção com a base de Sócios que receberam Seguro Defeso:
df = pd.read_csv(r'C:\DATASETS\Empresas Rotuladas\socios_empresas_rotuladas-seguro_defeso.csv', dtype=str, encoding='UTF-8')
socios_empresas_rotuladas_df = socios_empresas_rotuladas_df.merge(df, on=['cnpj', 'cnpj_cpf_socio', 'nome_socio', 'cnpj_cpf_socio_mascarado'], how='left')
# socio_solicitou_seguro_defeso
```

```python
# Junção com a base de Sócios que foram sancionados por Órgão de Controle:
df = pd.read_csv(r'C:\DATASETS\Empresas Rotuladas\socios_empresas_rotuladas-socio_sancionado.csv', dtype=str, encoding='UTF-8')
socios_empresas_rotuladas_df = socios_empresas_rotuladas_df.merge(df, on=['cnpj', 'cnpj_cpf_socio', 'nome_socio', 'cnpj_cpf_socio_mascarado'], how='left')
# socio_sancionado
```

### Conferindo resultado:

```python
socios_empresas_rotuladas_df
```

```python
print("Auxílio Emergencial:")
print(socios_empresas_rotuladas_df.socio_solicitou_auxilio_emergencial.value_counts())
print(socios_empresas_rotuladas_df.auxilio_emergencial_bloqueado.value_counts())
print("Bolsa Família:")
print(socios_empresas_rotuladas_df.socio_solicitou_bolsa_familia.value_counts())
print("Candidato:")
print(socios_empresas_rotuladas_df.socio_candidato.value_counts())
print("Devedor da União:")
socios_empresas_rotuladas_df.socio_devedor_uniao = np.where(socios_empresas_rotuladas_df.socio_devedor_uniao=="True",1,0)
print(socios_empresas_rotuladas_df.socio_devedor_uniao.value_counts())
print("Participação em Empresas Inexistentes:")
print(socios_empresas_rotuladas_df.empresas_inexistentes_socio_participa.value_counts())
print("Participação em Outros CNPJs:")
print(socios_empresas_rotuladas_df.qte_cnpj_socio_participa.value_counts())
print("Pessoa Politicamente Exposta:")
print(socios_empresas_rotuladas_df.socio_pessoa_politicamente_exposta.value_counts())
print("Recursos Recebidos do Governo Federal:")
socios_empresas_rotuladas_df.socio_recurso_recebido_governo_federal = socios_empresas_rotuladas_df.socio_recurso_recebido_governo_federal.astype(float)
print(socios_empresas_rotuladas_df[socios_empresas_rotuladas_df.socio_recurso_recebido_governo_federal>0].shape)
print(socios_empresas_rotuladas_df.socio_recurso_recebido_governo_federal.sum())
print("Seguro Defeso:")
print(socios_empresas_rotuladas_df.socio_solicitou_seguro_defeso.value_counts())
print("Sancionados:")
socios_empresas_rotuladas_df.socio_sancionado = np.where(socios_empresas_rotuladas_df.socio_sancionado=="True",1,0)
print(socios_empresas_rotuladas_df.socio_sancionado.value_counts())
```

```python
# Salvando resultado:
socios_empresas_rotuladas_df.to_csv(r'C:\DATASETS\Empresas Rotuladas\socios_empresas_rotuladas-consolidado_por_socio.csv', index=False)
```

```python
consolidado_df = socios_empresas_rotuladas_df.groupby(['cnpj']).agg({'cnpj_cpf_socio': 'count', 'tempo_de_sociedade': 'max', 'auxilio_emergencial_bloqueado': 'max', 'socio_solicitou_auxilio_emergencial': 'max', 'socio_solicitou_bolsa_familia': 'max', 'socio_candidato': 'max', 'socio_devedor_uniao': 'max', 'empresas_inexistentes_socio_participa': 'max', 'qte_cnpj_socio_participa': 'max', 'socio_pessoa_politicamente_exposta': 'max', 'socio_recurso_recebido_governo_federal': 'sum', 'socio_solicitou_seguro_defeso': 'max', 'socio_sancionado': 'max'}).reset_index()
consolidado_df
```

```python
consolidado_df.rename(columns={'cnpj_cpf_socio': 'qte_socios'}, inplace=True)
```

```python
print("Auxílio Emergencial:")
print(consolidado_df.socio_solicitou_auxilio_emergencial.value_counts())
print(consolidado_df.auxilio_emergencial_bloqueado.value_counts())
print("Bolsa Família:")
print(consolidado_df.socio_solicitou_bolsa_familia.value_counts())
print("Candidato:")
print(consolidado_df.socio_candidato.value_counts())
print("Devedor da União:")
print(consolidado_df.socio_devedor_uniao.value_counts())
print("Participação em Empresas Inexistentes:")
print(consolidado_df.empresas_inexistentes_socio_participa.value_counts())
print("Participação em Outros CNPJs:")
print(consolidado_df.qte_cnpj_socio_participa.value_counts())
print("Pessoa Politicamente Exposta:")
print(consolidado_df.socio_pessoa_politicamente_exposta.value_counts())
print("Recursos Recebidos do Governo Federal:")
print(consolidado_df[consolidado_df.socio_recurso_recebido_governo_federal>0].shape)
print(consolidado_df.socio_recurso_recebido_governo_federal.sum())
print("Seguro Defeso:")
print(consolidado_df.socio_solicitou_seguro_defeso.value_counts())
print("Sancionados:")
print(consolidado_df.socio_sancionado.value_counts())
```

```python
# Salvando resultado:
consolidado_df.to_csv(r'C:\DATASETS\Empresas Rotuladas\socios_empresas_rotuladas-consolidado_por_cnpj.csv', index=True)
```
