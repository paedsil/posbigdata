```python
import pandas as pd
import numpy as np
```

```python
# IMPORTAÇÃO DA BASE DE SÓCIOS
socios_empresas_rotuladas_df = pd.read_csv(r'C:\DATASETS\Empresas Rotuladas\socios_empresas_rotuladas.csv', dtype=str)
socios_empresas_rotuladas_df
```

```python
# Remover coluna de tempo_de_sociedade, não necessária aqui
socios_empresas_rotuladas_df.drop(columns=['tempo_de_sociedade'], inplace=True)
```

### Junção com a base de auxílio emergencial:

```python
# IMPORTAÇÃO DA BASE DE PESSOAS QUE FORAM CANDIDATAS NAS ELEIÇÕES:
auxilio_df = pd.read_csv(r'C:\DATASETS\Auxilio Emergencial\auxilio_emergencial.csv', dtype=str, encoding='latin-1')
auxilio_df
```

```python
socios_empresas_rotuladas_df = socios_empresas_rotuladas_df.merge(auxilio_df, left_on=['cnpj_cpf_socio_mascarado', 'nome_socio'], right_on=['CPF BENEFICIÁRIO', 'NOME BENEFICIÁRIO'], how='left', indicator='socio_solicitou_auxilio_emergencial')
socios_empresas_rotuladas_df.socio_solicitou_auxilio_emergencial = np.where(socios_empresas_rotuladas_df.socio_solicitou_auxilio_emergencial=='both', 1, 0).astype(np.uint8)
socios_empresas_rotuladas_df.auxilio_emergencial_bloqueado = np.where(socios_empresas_rotuladas_df.auxilio_emergencial_bloqueado=='True', 1, 0).astype(np.uint8)
socios_empresas_rotuladas_df.fillna({'socio_solicitou_auxilio_emergencial': 0, "auxilio_emergencial_bloqueado": 0}, inplace=True)
socios_empresas_rotuladas_df
```

Verificando resultados:

```python
print(socios_empresas_rotuladas_df.socio_solicitou_auxilio_emergencial.value_counts())
print(socios_empresas_rotuladas_df.auxilio_emergencial_bloqueado.value_counts())
```

```python
socios_empresas_rotuladas_df = socios_empresas_rotuladas_df.drop(columns=['CPF BENEFICIÁRIO', 'NOME BENEFICIÁRIO'])
```

```python
# Salvando resultado:
socios_empresas_rotuladas_df.to_csv(r'C:\DATASETS\Empresas Rotuladas\socios_empresas_rotuladas-auxilio_emergencial.csv', index=False)
```
