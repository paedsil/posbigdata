### Feature engineering:

##### Coluna 'email':

O e-mail por si só não é interessante para essa análise. Mas há duas informações relevantes que podemos extrair da coluna de e-mail: 1) se o e-mail foi ou não preenchido; 2) e qual é o domínio do e-mail. Um domínio específico (como bb.com.br, santanense.com.br) indica uma empresa mais estruturada por trás (mesmo que seja o e-mail da contabilidade). Já um domínio genérico (gmail.com, hotmail.com) não é tão forte nesse sentido. <br>Então, vamos extrair os domínios dos emails, e preencher com NI para os nulos. Vamos considerar as duas informações em uma mesma coluna (vamos criar a coluna 'email_processado').

```python
# Correção de registros como '@@hotmail.com':
empresas_df["email"] = empresas_df["email"].str.replace('@@', '@', regex=False)
# Criação da coluna email_processado>
empresas_df["email_processado"] = np.where((empresas_df["email"].astype(str).str.contains("@"))&(~empresas_df["email"].astype(str).str.endswith("@")), empresas_df["email"].astype(str).str.split("@").str[1], np.where(empresas_df["email"].isnull(), 'NI', empresas_df["email"]))
```

```python
empresas_df[["email","email_processado"]].describe
```

```python
# Vamos ver o resultado:
empresas_df["email_processado"] = empresas_df["email_processado"].str.upper()
empresas_df["email_processado"].value_counts()
```

Podemos remover a coluna de email original:

```python
empresas_df = empresas_df.drop(columns=['email'])
```

##### Colunas ddd_1 e telefone_1:

Por si só, não são relevantes, mas vamos extrair delas duas informações: 1) se o telefone foi preenchido ou não; 2) se o telefone é utilizado por outra empresa.

Primeiro, vamos ver se o telefone é utilizado por mais de uma empresa:

```python
# Renomear a coluna:
telefones_df = telefones_df.rename(columns={'cnpj': 'qte_empresas_com_mesmo_telefone'})
```

```python
empresas_df = empresas_df.merge(telefones_df, on='telefone_processado', how='left')
empresas_df
```

##### Colunas de endereços:

As colunas de endereço por si só não são interessantes para essa análise. Mas vamos extrair delas uma informação importante: quantas empresas compartilham do exato mesmo endereço? (pode acontecer, mas é um indicador de suspeita, prática comum em empresas fictícias ou montadas para algum propósito irregular, como em fraudes de licitações).

```python
# Colunas de endereço:
cols = ['tipo_logradouro', 'logradouro', 'numero', 'complemento', 'bairro', 'municipio', 'uf']

# Primeiro, fazemos preparação das colunas de enderaço para o join e posterior merge:
empresas_df = empresas_df.fillna(value={'tipo_logradouro':'', 'logradouro':'', 'numero':'', 'complemento':'', 'bairro':'', 'municipio':'', 'uf':''})
for col in cols:
    empresas_df[col] = empresas_df[col].str.upper()
    
# Concatenar as colunas de endereço:
empresas_df['endereco_processado'] = empresas_df[cols].apply(lambda row: '_'.join(row.values.astype(str)), axis=1)
empresas_df['endereco_processado'].describe
```

```python
# Calcular quantas empresas compartilham do mesmo endereço:
enderecos_df = empresas_df[['cnpj','endereco_processado']].groupby(['endereco_processado'])['cnpj'].count().reset_index()
enderecos_df = enderecos_df.rename(columns={'cnpj': 'qte_empresas_com_mesmo_endereco'})
enderecos_df
```

##### Colunas que vieram da base de sócios:

Sobre as colunas que vieram da base de sócios e que ficaram com valores nulos, são empresas para as quais não identificamos os sócios. Vamos preencher os valores com base nas estatísticas da base (média ou moda, conforme cada coluna):

```python
# Calculando a moda das colunas:
modas_df = empresas_df[['qte_socios', 'tempo_de_sociedade', 'auxilio_emergencial_bloqueado', 
                        'socio_solicitou_auxilio_emergencial', 'socio_solicitou_bolsa_familia',
                        'socio_candidato', 'socio_devedor_uniao', 'empresas_inexistentes_socio_participa',
                        'qte_cnpj_socio_participa', 'socio_pessoa_politicamente_exposta',
                        'socio_solicitou_seguro_defeso', 'socio_sancionado']].mode()
modas_df
```

```python
# Preenchendo os nulos:
empresas_df = empresas_df.fillna(value={
    'qte_socios': modas_df.qte_socios[0],
    'tempo_de_sociedade': modas_df.tempo_de_sociedade[0],
    'auxilio_emergencial_bloqueado': modas_df.auxilio_emergencial_bloqueado[0],
    'socio_solicitou_auxilio_emergencial': modas_df.socio_solicitou_auxilio_emergencial[0],
    'socio_solicitou_bolsa_familia': modas_df.socio_solicitou_bolsa_familia[0],
    'socio_candidato': modas_df.socio_candidato[0],
    'socio_devedor_uniao': modas_df.socio_devedor_uniao[0],
    'empresas_inexistentes_socio_participa': modas_df.empresas_inexistentes_socio_participa[0],
    'qte_cnpj_socio_participa': modas_df.qte_cnpj_socio_participa[0],
    'socio_pessoa_politicamente_exposta': modas_df.socio_pessoa_politicamente_exposta[0],
    'socio_recurso_recebido_governo_federal': 0,
    'socio_solicitou_seguro_defeso': modas_df.socio_solicitou_seguro_defeso[0],
    'socio_sancionado': modas_df.socio_sancionado[0]})

# Obs: para 'socio_recurso_recebido_governo_federal', seria usada a média,
# mas não foi possível calculá-la. O notebook sempre travava. Foi assumido o valor 0 para os nulos.
```

```python
empresas_df.qte_socios = empresas_df.qte_socios.astype(int)
empresas_df.tempo_de_sociedade = empresas_df.tempo_de_sociedade.astype(int)
empresas_df.auxilio_emergencial_bloqueado = empresas_df.auxilio_emergencial_bloqueado.astype(int)
empresas_df.socio_solicitou_auxilio_emergencial = empresas_df.socio_solicitou_auxilio_emergencial.astype(int)
empresas_df.socio_solicitou_bolsa_familia = empresas_df.socio_solicitou_bolsa_familia.astype(int)
empresas_df.socio_candidato = empresas_df.socio_candidato.astype(int)
empresas_df.socio_devedor_uniao = empresas_df.socio_devedor_uniao.astype(int)
empresas_df.empresas_inexistentes_socio_participa = empresas_df.empresas_inexistentes_socio_participa.astype(int)
empresas_df.qte_cnpj_socio_participa = empresas_df.qte_cnpj_socio_participa.astype(int)
empresas_df.socio_pessoa_politicamente_exposta = empresas_df.socio_pessoa_politicamente_exposta.astype(int)
empresas_df.socio_recurso_recebido_governo_federal = empresas_df.socio_recurso_recebido_governo_federal.astype(float)
empresas_df.socio_solicitou_seguro_defeso = empresas_df.socio_solicitou_seguro_defeso.astype(int)
empresas_df.socio_sancionado = empresas_df.socio_sancionado.astype(int)
```
