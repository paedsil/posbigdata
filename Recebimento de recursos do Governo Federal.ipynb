```python
import pandas as pd
import numpy as np
import glob
```

```python
fields = ["Código Favorecido", "Nome Favorecido", "Valor Recebido"]
fields_dtype = {"Código Favorecido": str, "Nome Favorecido": str, "Valor Recebido": str}
```

```python
# IMPORTAÇÃO DOS ARQUIVOS CSV DOS FAVORECIDOS DE RECURSO DO GOVERNO FEDERAL

path = r'C:\DATASETS\Recebimento Recursos Governo Federal\1-Arquivos Baixados'
all_files = glob.glob(path + "/*.csv")

li = []

for filename in all_files:
    temp_df = pd.read_csv(filename, index_col=None, header=0, sep=';', 
                          encoding='ANSI', dtype=fields_dtype, usecols=fields)
    temp_df['Arquivo'] = filename.rsplit('\\',1)[1]
    li.append(temp_df)

# Concatena o resultado no dataframe
favorecidos_df = pd.concat(li, axis=0, ignore_index=True)

favorecidos_df.describe
```

```python
favorecidos_df.Arquivo.value_counts()
```

```python
favorecidos_df['Valor Recebido'] = favorecidos_df['Valor Recebido'].str.replace(',', '.').astype(float)
```

```python
favorecidos_df
```

```python
# Salvando resultado completo:
favorecidos_df.to_csv(r'C:\DATASETS\Recebimento Recursos Governo Federal\recebimento_recursos_governo-completo.csv', index = False)
```

Todos os casos acima podem ser removidos. Vamos ficar apenas com CPF e CNPJ:

```python
favorecidos_df = favorecidos_df[favorecidos_df['Código Favorecido'].str.len()==14]
favorecidos_df['Código Favorecido'].str.len().value_counts()
```

```python
# Limpa os CPFs:
favorecidos_df['Código Favorecido'] = np.where(favorecidos_df['Código Favorecido'].str.startswith('*'), 
                        favorecidos_df['Código Favorecido'].str.replace('[\.\-]', '', regex=True),
                        favorecidos_df['Código Favorecido'])
favorecidos_df['Código Favorecido'].str.len().value_counts()
```

Quantidades conforme cálculo feito anteriormente.

```python
favorecidos_df
```

Vamos consolidar os valores:

```python
consolidado_df = favorecidos_df.groupby(['Código Favorecido', 'Nome Favorecido']).agg({'Valor Recebido': 'sum'})
consolidado_df
```

```python
consolidado_df[consolidado_df['Valor Recebido'] < 0]
```

Os valores negativos pesquisados se referem a estornos / cancelamentos. Vamos deixá-los no dataframe.

```python
consolidado_df[consolidado_df['Valor Recebido'] == 0]
```

Os valores zerados pesquisados são devidos a estornos / cancelamentos. Vamos deixá-los na base.

```python
consolidado_df
```

```python
consolidado_df.rename(columns={'Código Favorecido': 'CPF_CNPJ', 'Nome Favorecido': 'nome_Favorecido', 'Valor Recebido': 'recebimento_recurso_governo_federal'}, inplace=True)
```

```python
# Salvando resultado final:
consolidado_df.to_csv(r'C:\DATASETS\Recebimento Recursos Governo Federal\recebimento_recursos_governo-consolidado.csv', index = True)
```
