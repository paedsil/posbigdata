```python
import pandas as pd
import numpy as np
import re
```

```python
# IMPORTAÇÃO DA BASE DE SÓCIOS
socios_df = pd.read_csv('C:\\DATASETS\\CNPJ\\2-processado\\socios.csv', sep=',', dtype=str, usecols=["cnpj", "nome_socio", "cnpj_cpf_socio"])
```

```python
# Conferindo campos e valores
socios_df.describe
```

```python
# IMPORTAÇÃO DA BASE DE SOCIOS DAS EMPRESAS ROTULADAS
socios_empresas_rotuladas_df = pd.read_csv(r'C:\DATASETS\Empresas Rotuladas\socios_empresas_rotuladas.csv', sep=',', dtype=str, usecols=['cnpj','cnpj_cpf_socio','nome_socio','cnpj_cpf_socio_mascarado'])
socios_empresas_rotuladas_df
```

```python
socios_empresas_rotuladas_df2 = socios_empresas_rotuladas_df[['cnpj_cpf_socio_mascarado', 'nome_socio']]
socios_empresas_rotuladas_df2 = socios_empresas_rotuladas_df2.drop_duplicates()
socios_empresas_rotuladas_df2
```

```python
socios_df2 = socios_df.merge(socios_empresas_rotuladas_df2, left_on=['nome_socio','cnpj_cpf_socio'], right_on=['nome_socio','cnpj_cpf_socio_mascarado'], how='inner')
socios_df2
```

```python
outras_empresas_df = socios_df2.groupby(['cnpj_cpf_socio_mascarado', 'nome_socio'])['cnpj']
                          .count().reset_index()
outras_empresas_df
```

```python
outras_empresas_df = outras_empresas_df.rename(columns={"cnpj": "qte_cnpj_socio_participa"})
```

```python
outras_empresas_df[outras_empresas_df.qte_cnpj_socio_participa > 1].shape
```

156.840 sócios de empresas rotuladas participam de mais de 1 CNPJ.

```python
# Rotula os sócios
socios_empresas_rotuladas_df = socios_empresas_rotuladas_df.merge(outras_empresas_df, on=['cnpj_cpf_socio_mascarado', 'nome_socio'], how='left')
socios_empresas_rotuladas_df
```

Os valores NaN na coluna qte_cnpj_socio_participa são das empresas que estão sem sócios na base da RFB, mas que foi possível identificar o sócio a partir da extração do campo razão social. Vamos atribuir a eles o valor 1:

```python
socios_empresas_rotuladas_df.qte_cnpj_socio_participa.isnull().sum()
```

```python
socios_empresas_rotuladas_df.qte_cnpj_socio_participa.fillna(1, inplace=True)
socios_empresas_rotuladas_df.qte_cnpj_socio_participa = socios_empresas_rotuladas_df.qte_cnpj_socio_participa.astype(int)
socios_empresas_rotuladas_df
```
