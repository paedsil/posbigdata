```python
import pandas as pd
import numpy as np
```

```python
# IMPORTAÇÃO DA BASE DE EMPRESAS ROTULADAS
empresas_rotuladas_df = pd.read_csv(r'C:\DATASETS\Empresas Rotuladas\empresas_rotuladas.csv', dtype=str, usecols=['cnpj'])
empresas_rotuladas_df
```

### Junção com a base de devedores da dívida ativa da União:

```python
# IMPORTAÇÃO DA BASE DE DÍVIDA ATIVA
divida_ativa_df = pd.read_csv(r'C:\DATASETS\Divida_Ativa\divida_ativa_todos.csv', dtype=str)
divida_ativa_df
```

```python
# Só precisamos dos CNPJs:
divida_ativa_df = divida_ativa_df[divida_ativa_df.CPF_CNPJ.str.len()==14]
```

```python
# Rotula as empresas
empresas_rotuladas_df['empresa_devedora_uniao'] = (empresas_rotuladas_df.merge(divida_ativa_df, left_on='cnpj', right_on='CPF_CNPJ', how='left', indicator=True).eval('_merge == "both"').astype(bool))
empresas_rotuladas_df.empresa_devedora_uniao.value_counts()
```
