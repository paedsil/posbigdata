### Vamos agora acessar cada página e extrair os CNPJs:

```python
pattern = re.compile('\D(\d?\d\D?\d{3}\D?\d{3}\D?\d{4}\s?\D?\s?\d{2})')
empresas_baixadas_df = pd.DataFrame(columns=['CNPJ', 'ADE', 'Link'])
# contador = 0
for link in nova_lista_link_ADE:
    print('Acessando: ', link)
    
    # Para contornar problemas de conexão:
    for tentativa in range(3):
        try:
            response = requests.get(link, headers={"User-Agent" : "Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/51.0.2704.103 Safari/537.36"})
        except Exception as e:
            print('Erro de conexão (', tentativa, '): ', e)
    
    if not response:
        print('  > LINK PULADO!!!!')
        continue
    
    soup = BeautifulSoup(response.text, 'html.parser')
    
    # Obtém o texto do ADE e faz algumas verificações:
    texto = soup.find("body").text.upper()
    
    # Verifica se o ADE possui alguma variação do texto INEXISTE DE FATO:
    if not (re.search('INEXIST.{1,6} DE FATO', texto)):
        # Como não encontrou o texto, verifica se há menção aos normativos que regulam a baixa de ofício por inexistência de fato
        if (re.search('(INSTRUÇÃO NORMATIVA|IN).+(1.?863.+?2.?018|1.?634.+?2.?016)', texto) == None) or (re.search('ART.{1,6}(29|31)', texto) == None):
            # Verifica se é um ato que não tem mais efeito, ou se é de baixa de CPF:
            m = re.search('(ANULADO|REVOGAGO|SUSPENSO|SEM EFEITO|CPF)', texto)
            print('  > Não encontrou inexistente de fato e nem IN 1.863/2018 ou 1.634/2016!', m.groups() if m else '')
            continue
    
    # Verifica se o ato efetivamente 'declara'
    if re.search("(DECLAR.{1,5})", texto) == None:
        print('  > Não encontrou o termo DECLARA!')
        continue
    
    cnpjs_encontrados = pattern.findall(texto)
    
    qte_cnpjs_encontrados = len(cnpjs_encontrados)
    
    if qte_cnpjs_encontrados == 0:
        m = re.search('(ANULAD.|REVOGAD.|SUSPENS.|SEM EFEITO|CPF)', texto)
        print('  > Nenhum CNPJ encontrado!', m.groups() if m else '')
        continue
            
    print('  > CNPJs encontrados: ', qte_cnpjs_encontrados)
    
    # Obtém o título do ADE
    ADE = soup.find("div", {"class": "tituloAto"}).text
    ADE = re.sub(r'[\t\n]', '', ADE)
    ADE = re.sub(r'\s+', ' ', ADE)
    
    temp_df = pd.DataFrame(columns=['CNPJ', 'ADE', 'Link'])
    temp_df['CNPJ'] = cnpjs_encontrados
    temp_df['ADE'] = ADE
    temp_df['Link'] = link

    empresas_baixadas_df = pd.concat([empresas_baixadas_df, temp_df], ignore_index=True)    
    
empresas_baixadas_df
```

Vamos validar os CNPJS:

```python
# Função de validar CNPJ, copiada de https://github.com/rafahlobo/cpfValidator/blob/master/cpf.py
def isCnpjValid(cnpj):
    """ If cnpf in the Brazilian format is valid, it returns True, otherwise, it returns False. """

    # Check if type is str
    if not isinstance(cnpj,str):
        return False

    # Remove some unwanted characters
    cpf = re.sub("[^0-9]",'',cnpj)

    # Checks if string has 11 characters
    if len(cpf) != 14:
        return False

    sum = 0
    weight = [5,4,3,2,9,8,7,6,5,4,3,2]

    """ Calculating the first cpf check digit. """
    for n in range(12):
        value =  int(cpf[n]) * weight[n]
        sum = sum + value


    verifyingDigit = sum % 11

    if verifyingDigit < 2 :
        firstVerifyingDigit = 0
    else:
        firstVerifyingDigit = 11 - verifyingDigit

    """ Calculating the second check digit of cpf. """
    sum = 0
    weight = [6,5,4,3,2,9,8,7,6,5,4,3,2]
    for n in range(13):
        sum = sum + int(cpf[n]) * weight[n]

    verifyingDigit = sum % 11

    if verifyingDigit < 2 :
        secondVerifyingDigit = 0
    else:
        secondVerifyingDigit = 11 - verifyingDigit

    if cpf[-2:] == "%s%s" % (firstVerifyingDigit,secondVerifyingDigit):
        return True
    return False
```

```python
# Remove os CNPJs inválidos:
empresas_baixadas_df = empresas_baixadas_df.drop(empresas_baixadas_df[empresas_baixadas_df.CNPJ.apply(isCnpjValid) == False].index)
empresas_baixadas_df
```
