```python
import pandas as pd
import numpy as np
```

```python
# IMPORTAÇÃO DA BASE DE SÓCIOS
socios_empresas_rotuladas_df = pd.read_csv(r'C:\DATASETS\Empresas Rotuladas\socios_empresas_rotuladas.csv', dtype=str)
socios_empresas_rotuladas_df
```

```python
# Remover coluna de tempo_de_sociedade, não necessária aqui
socios_empresas_rotuladas_df.drop(columns=['tempo_de_sociedade'], inplace=True)
```

### Junção com a base do bolsa família:

```python
# IMPORTAÇÃO DA BASE DE PESSOAS QUE RECEBERAM PAGAMENTO DO BOLSA FAMÍLIA:
bolsa_familia_df = pd.read_csv(r'C:\DATASETS\Bolsa Família\bolsa_familia.csv', dtype=str, encoding='latin-1')
bolsa_familia_df
```

```python
socios_empresas_rotuladas_df['socio_solicitou_bolsa_familia'] = socios_empresas_rotuladas_df.merge(bolsa_familia_df, left_on=['cnpj_cpf_socio_mascarado', 'nome_socio'], right_on=['CPF', 'nome_favorecido'], how='left', indicator=True).eval('_merge == "both"').astype(int)
socios_empresas_rotuladas_df
```

Verificando resultados:

```python
socios_empresas_rotuladas_df.socio_solicitou_bolsa_familia.value_counts()
```

```python
# Salvando resultado:
socios_empresas_rotuladas_df.to_csv(r'C:\DATASETS\Empresas Rotuladas\socios_empresas_rotuladas-bolsa_familia.csv', index=False)
```
