```python
import pandas as pd
import numpy as np
```

```python
# IMPORTAÇÃO DA BASE DE SÓCIOS
socios_empresas_rotuladas_df = pd.read_csv(r'C:\DATASETS\Empresas Rotuladas\socios_empresas_rotuladas.csv', dtype=str)
socios_empresas_rotuladas_df
```

```python
# Remover coluna de tempo_de_sociedade, não necessária aqui
socios_empresas_rotuladas_df.drop(columns=['tempo_de_sociedade'], inplace=True)
```

### Junção com a base de candidatos:

```python
# IMPORTAÇÃO DA BASE DE PESSOAS QUE FORAM CANDIDATAS NAS ELEIÇÕES:
candidatos_df = pd.read_csv(r'C:\DATASETS\Candidatos\candidatos.csv', dtype=str, encoding='latin-1')
candidatos_df
```

```python
candidatos_df['cpf_mascarado'] = '***' + candidatos_df.CPF.str[3:9] + '**'
```

```python
candidatos_df
```

```python
socios_empresas_rotuladas_df['socio_candidato'] = socios_empresas_rotuladas_df.merge(candidatos_df, left_on=['cnpj_cpf_socio_mascarado', 'nome_socio'], right_on=['cpf_mascarado', 'nome_candidato'], how='left', indicator=True).eval('_merge == "both"').astype(int)
socios_empresas_rotuladas_df
```

Verificando resultados:

```python
socios_empresas_rotuladas_df.socio_candidato.value_counts()
```

```python
# Salvando resultado:
socios_empresas_rotuladas_df.to_csv(r'C:\DATASETS\Empresas Rotuladas\socios_empresas_rotuladas-candidato.csv', index = False)
```
