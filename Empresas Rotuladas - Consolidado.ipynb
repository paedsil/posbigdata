### 3. Processamento/Tratamento de Dados

```python
import pandas as pd
import numpy as np
```

```python
# Mostrar todas as colunas dos dataframes
pd.set_option('display.max_columns', None)
```

```python
# IMPORTAÇÃO DA BASE DE EMPRESAS
empresas_rotuladas_df = pd.read_csv(r'C:\DATASETS\Empresas Rotuladas\empresas_rotuladas.csv', dtype=str)
empresas_rotuladas_df
```

### Junções com as bases de PJ:

```python
# Junção com a base de atendimentos dos Procons:
df = pd.read_csv(r'C:\DATASETS\Empresas Rotuladas\empresas_rotuladas-atendimentos_procon.csv', dtype=str, encoding='UTF-8')
empresas_rotuladas_df = empresas_rotuladas_df.merge(df, on='cnpj', how='left')
# Atendimentos_Procon
```

```python
# Junção com a base de Devedores da União:
df = pd.read_csv(r'C:\DATASETS\Empresas Rotuladas\empresas_rotuladas-devedores_da_uniao.csv', dtype=str, encoding='UTF-8')
empresas_rotuladas_df = empresas_rotuladas_df.merge(df, on='cnpj', how='left')
# empresa_devedora_uniao
```

```python
# Junção com a base de empresas que participaram de licitação:
df = pd.read_csv(r'C:\DATASETS\Empresas Rotuladas\empresas_rotuladas-participa_licitacao.csv', dtype=str, encoding='UTF-8')
empresas_rotuladas_df = empresas_rotuladas_df.merge(df, on='cnpj', how='left')
# empresa_venceu_licitacao,empresa_participou_licitacao
```

```python
# Junção com a base de empresas que receberam recursos da união:
df = pd.read_csv(r'C:\DATASETS\Empresas Rotuladas\empresas_rotuladas-recebimento_recurso_governo_federal.csv', dtype=str, encoding='UTF-8')
empresas_rotuladas_df = empresas_rotuladas_df.merge(df, on='cnpj', how='left')
# empresa_recurso_recebido_governo_federal
```

```python
# Junção com a base de empresas sancionadas por órgão de controle:
df = pd.read_csv(r'C:\DATASETS\Empresas Rotuladas\empresas_rotuladas-sancionadas.csv', dtype=str, encoding='UTF-8')
empresas_rotuladas_df = empresas_rotuladas_df.merge(df, on='cnpj', how='left')
# empresa_sancionada
```

##### Conferindo resultado das junções:

```python
empresas_rotuladas_df = empresas_rotuladas_df.drop_duplicates()
```

```python
empresas_rotuladas_df
```

```python
print("Atendimentos Procon's:")
print(empresas_rotuladas_df.Atendimentos_Procon.value_counts())
print("Devedor da União:")
empresas_rotuladas_df.empresa_devedora_uniao = np.where(empresas_rotuladas_df.empresa_devedora_uniao=="True",1,0)
print(empresas_rotuladas_df.empresa_devedora_uniao.value_counts())
print("Participaram de licitação:")
print(empresas_rotuladas_df.empresa_participou_licitacao.value_counts())
print(empresas_rotuladas_df.empresa_venceu_licitacao.value_counts())
print("Recursos Recebidos do Governo Federal:")
empresas_rotuladas_df.empresa_recurso_recebido_governo_federal = empresas_rotuladas_df.empresa_recurso_recebido_governo_federal.astype(float)
print(empresas_rotuladas_df[empresas_rotuladas_df.empresa_recurso_recebido_governo_federal>0].shape)
print(empresas_rotuladas_df.empresa_recurso_recebido_governo_federal.sum())
print("Sancionados:")
empresas_rotuladas_df.empresa_sancionada = np.where(empresas_rotuladas_df.empresa_sancionada=="True",1,0)
print(empresas_rotuladas_df.empresa_sancionada.value_counts())
```

### Junções com a base de Sócios consolidada:

```python
# Junção com a base de sócios consolidada por cnpj:
df = pd.read_csv(r'C:\DATASETS\Empresas Rotuladas\socios_empresas_rotuladas-consolidado_por_cnpj.csv', dtype=str, encoding='UTF-8')
empresas_rotuladas_df = empresas_rotuladas_df.merge(df, on='cnpj', how='left')
```
