```python
import pandas as pd
import numpy as np
```

```python
# IMPORTAÇÃO DA BASE DE SÓCIOS
socios_empresas_rotuladas_df = pd.read_csv(r'C:\DATASETS\Empresas Rotuladas\socios_empresas_rotuladas.csv', dtype=str)
socios_empresas_rotuladas_df
```

```python
# Remover coluna de tempo_de_sociedade, não necessária aqui
socios_empresas_rotuladas_df.drop(columns=['tempo_de_sociedade'], inplace=True)
```

### Junção com a base de pessoas politicamente expostas:

```python
# IMPORTAÇÃO DA BASE DE PESSOAS POLITICAMENTE EXPOSTAS:
pep_df = pd.read_csv(r'C:\DATASETS\Pessoa Exposta Politicamente\pessoa_politicamente_exposta.csv', dtype=str, encoding='latin-1')
pep_df
```

```python
socios_empresas_rotuladas_df['socio_pessoa_politicamente_exposta'] = socios_empresas_rotuladas_df.merge(pep_df, left_on=['cnpj_cpf_socio_mascarado', 'nome_socio'], right_on=['CPF', 'Nome_PEP'], how='left', indicator=True).eval('_merge == "both"').astype(int)
socios_empresas_rotuladas_df
```

Verificando resultados:

```python
socios_empresas_rotuladas_df.socio_pessoa_politicamente_exposta.value_counts()
```

```python
# Salvando resultado:
socios_empresas_rotuladas_df.to_csv(r'C:\DATASETS\Empresas Rotuladas\socios_empresas_rotuladas-pessoa_politicamente_exposta.csv', index = False)
```
